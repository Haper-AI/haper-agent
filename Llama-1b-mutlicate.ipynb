```python
from huggingface_hub import snapshot_download

# 下载 LLaMA-1B
model_id = "meta-llama/Llama-3.2-1B"  # 确保 LLaMA-1B 可用，否则换成 LLaMA-7B 或更小的模型
model_path = "./llama-1b"

snapshot_download(repo_id=model_id, local_dir=model_path, resume_download=True)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# ==== Step 1: 加载模型 ====
model_name = "meta-llama/Llama-3.2-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# ==== Step 2: 读取 LLaMA 格式 txt 文件 ====
llama_txt_path = "/content/drive/MyDrive/AI agent/llama1b_email_classification_prompts.txt"  # 改为你的实际路径

with open(llama_txt_path, "r", encoding="utf-8") as f:
    content = f.read()

raw_samples = [s.strip() for s in content.split("</s>") if s.strip()]

# ==== Step 3: 提取 prompt 和真实标签 ====
def extract_prompt_and_label(text):
    try:
        if "Answer:" not in text:
            return None, None
        prompt, label = text.split("Answer:", 1)
        return prompt.strip() + "\nAnswer:", label.strip()
    except Exception:
        return None, None

prompts = []
true_labels = []

for s in raw_samples:
    prompt, label = extract_prompt_and_label(s)
    if prompt and label:
        prompts.append(prompt)
        true_labels.append(label)

print(f"📦 成功解析样本数：{len(prompts)}")
print("✅ 示例 prompt：", prompts[0])
print("✅ 标签：", true_labels[0])

# === Step 4: 模型推理 ===
predicted_labels = []
raw_outputs = []

for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512).to(model.device)
    input_len = inputs.input_ids.shape[1]

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=10,
            temperature=0.0,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    gen = outputs[0][input_len:]
    pred_text = tokenizer.decode(gen, skip_special_tokens=True).strip()
    raw_outputs.append(pred_text)

    # 提取预测 label（第一个词）
    prediction = pred_text.split()[0].strip().capitalize() if pred_text else "[EMPTY]"
    predicted_labels.append(prediction)

# === Step 5: 准确率计算 ===
valid_labels = {"Receipt", "Promotion", "Ads", "Notice", "Career", "Career Development", "Education", "News"}
true_clean = [label.strip().capitalize() for label in true_labels]
correct = sum(p == t and p in valid_labels for p, t in zip(predicted_labels, true_clean))
accuracy = correct / len(true_clean)

# === Step 6: 打印结果 ===
print(f"\n🎯 分类准确率: {accuracy * 100:.2f}%")
print("\n🔍 示例预测（前 5 条）：")
for i in range(min(5, len(prompts))):
    print("=" * 60)
    print(f"📨 Prompt:\n{prompts[i]}...\n")
    print(f"📤 模型输出: {raw_outputs[i]}")
    print(f"✅ True Label: {true_clean[i]} | 🔮 Predicted: {predicted_labels[i]}")
    print("=" * 60 + "\n")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# ==== Step 1: 加载模型 ====
model_name = "meta-llama/Llama-3.2-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# ==== Step 2: 读取 prompts 文件 ====
with open("/content/drive/MyDrive/AI agent/llama_email_action_prompts_no_operation.txt", "r", encoding="utf-8") as f:
    content = f.read()

samples = [s.strip() for s in content.split("</s>") if s.strip()]

# ==== Step 3: 正确提取 prompt 和 true label（从最后一个 Operation? 分割）====
prompts = []
true_labels = []

for s in samples:
    if "Operation?" in s:
        try:
            # 分成两部分：prompt 和 label
            prompt_part, label_part = s.rsplit("Operation?", 1)

            # 构造干净的 prompt，只保留到 Operation?
            prompt = prompt_part.strip() + "Operation?"

            # 提取 answer（去掉 Answer: 前缀，处理大小写）
            label = label_part.replace("Answer:", "").strip()

            prompts.append(prompt)
            true_labels.append(label)
        except:
            continue

print(f"📦 成功解析样本数：{len(prompts)}")
print("✅ 示例 prompt：", prompts[0])
print("✅ 标签：", true_labels[0])

# ==== Step 4: 推理 + 提取预测 ====
valid_ops = {"Browse", "Reply", "Delete", "Scrutinize"}

def extract_operation(pred_text):
    """从模型输出中提取有效操作名"""
    for word in pred_text.replace("\n", " ").split():
        candidate = word.strip().capitalize()
        if candidate in valid_ops:
            return candidate
    return "[INVALID]"

predicted = []
raw_outputs = []

for prompt in prompts:
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512).to(model.device)
    input_len = inputs.input_ids.shape[1]

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=8,
            temperature=0.0,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    gen_tokens = output[0][input_len:]
    pred_text = tokenizer.decode(gen_tokens, skip_special_tokens=True).strip()
    raw_outputs.append(pred_text)

    operation = extract_operation(pred_text)
    predicted.append(operation)

# ==== Step 5: 准确率评估 ====
true_clean = [label.strip().capitalize() for label in true_labels]
correct = sum(p == t and p in valid_ops for p, t in zip(predicted, true_clean))
accuracy = correct / len(true_clean) if true_clean else 0.0

# ==== Step 6: 打印预测结果 ====
print(f"\n✅ 样本总数: {len(prompts)}")
print(f"🎯 分类准确率: {accuracy * 100:.2f}%")

print("\n🔍 示例预测（前 5 条）：")
for i in range(min(5, len(prompts))):
    print("=" * 60)
    print(f"{i+1}. Prompt:\n{prompts[i][:300]}...")  # 长 prompt 做截断
    print(f"✅ True Label: {true_clean[i]}")
    print(f"🔮 Predicted : {predicted[i]}")
    print(f"📤 Raw Output: {raw_outputs[i]}\n")
```
