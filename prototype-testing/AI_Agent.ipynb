```python
prompt = f"""
Task: Classify an email into one of three categories: 'Important', 'Not Important', or 'Spam'.

Guidelines:
1. Classify an email as 'Important' if it contains content that is personally relevant to the recipient, such as:
   - Work-related messages
   - Financial updates
   - Urgent notifications
   - Personal correspondence
   - Scheduled events

2. Classify an email as 'Not Important' if it contains:
   - Promotional messages
   - Social media notifications
   - Newsletters
   - Advertisements
   - Any content that does not require immediate attention

3. Classify an email as 'Spam' if it exhibits any of the following characteristics:
   - Phishing attempts (requests for personal, financial, or login information)
   - Unsolicited promotions or marketing emails the recipient did not sign up for
   - Scam emails (e.g., lottery winnings, fake job offers, fraudulent financial opportunities)
   - Bulk marketing messages with generic content
   - Malicious content (suspicious links, attachments, or requests for sensitive data)

Ensure that spam emails are never classified as ‘Important.’ Prioritize accuracy by analyzing the sender, subject, and content.

Examples:
1. Email: "Hi, please submit your report by EOD today. The client is waiting for it."
   Classification: Important

2. Email: "Reminder: Team lunch tomorrow at 1 PM."
   Classification: Not Important

3. Email: "URGENT! Your bank account has been compromised. Click here to verify your details immediately."
   Classification: Spam

4. Email: "🔥Huge Discounts on Luxury Watches! Limited Time Offer!"
   Classification: Spam

5. Email: "Your PayPal account has been restricted. Verify your details now!"
   Classification: Spam

Email:
Subject: '{email_data['subject']}'
{email_data['body']}
{email_data['sender']}

Classification:
"""
```
